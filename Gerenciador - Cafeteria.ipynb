# Setup

```python
#Importacao de Bibliotecas
import csv
import time
import os
import platform
from IPython.display import clear_output
from datetime import date
from datetime import datetime

def carregar_arquivo(arquivo,delimitador):
    with open(arquivo, newline='', encoding='utf-8-sig') as f:
        return list(csv.DictReader(f, delimiter=delimitador))
```

## Cardapio

```python
def Cardapio_():
    caminho_csv = r"C:\Users\Desktop\Documents\Trabalho\Automação\Projeto\Auxiliar\Cardapio.csv"
    Cardapio = carregar_arquivo(caminho_csv,",")
    Categorias = set(item['Categoria'] for item in Cardapio)
    Categorias = sorted(list(Categorias))
    
    while True:
        try:
            print("Escolha a categoria desejada(1-5):")
            j = 0
            for i in Categorias:
                j+=1
                print(str(j)+"."+i)
            print("0. Sair")
        
            exibir = int(input("N° da Categoria: "))
            clear_output()

            if exibir == 0:
                print("Saindo...")
                break
                
            print("_______________\n")
            j = 0

            for i in range(len(Cardapio)):
                if Cardapio[i]['Categoria'] == Categorias[exibir-1]:
                    j+=1

                    if Cardapio[i]['Promocao'] == "sim":
                        print(Cardapio[i]['ID_Item']+"."+Cardapio[i]['Item']+" - Promoção de R$"+Cardapio[i]['Preco']+" por R$"+Cardapio[i]['PrecoPromocao']+"!\nIngredientes: "+Cardapio[i]['Ingredientes']+"\n")
                    else:
                        print(Cardapio[i]['ID_Item']+"."+Cardapio[i]['Item']+" - R$"+Cardapio[i]['Preco']+"\nIngredientes: "+Cardapio[i]['Ingredientes']+"\n")
            print("_______________")

        except (ValueError, IndexError):
            clear_output()
            print("Entrada inválida!\n")
```

## Clientes

```python
def Clientes_():
    caminho_csv = r"C:\Users\Desktop\Documents\Trabalho\Automação\Projeto\Auxiliar\Clientes.csv"

    while True:
        clear_output()

        print("Digite a opção desejada:"
            "\n1. Cadastrar um Cliente"
            "\n2. Consultar Lista de Clientes Cadastrados"
            "\n3. Descadastrar um Cliente"
            "\n0. Sair")

        try:
            opcao = int(input("Opção: "))
        except ValueError:
            print("Digite um número válido.")
            input("Pressione Enter para continuar...")
            continue

        Clientes = carregar_arquivo(caminho_csv,";")
        if(opcao == 1):
            clear_output()
            print("Cadastramento Iniciado\nDigite o nome do cliente a ser cadastrado:")
            nome = input("Digite o nome do cliente:")
            id = datetime.now().strftime("%Y%m%d%H%M%S")

            dit = {'Data_Cadastro': str(date.today()),
                    'ID': id,
                    'Nome': nome,
                    'Pontos': 0,
                    }

            with open(caminho_csv, mode='a', newline='', encoding='utf-8') as arquivo_csv:
                    campos = ['Data_Cadastro', 'ID', 'Nome', 'Pontos']
                    escritor = csv.DictWriter(arquivo_csv, fieldnames=campos, delimiter = ";")
                    escritor.writerow(dit)

            clear_output()
            print("Cliente: "+nome+"\nID: "+id+"\nCadastrado com Sucesso!")

        elif(opcao == 2):
            clear_output()
            print("Lista de Clientes Cadastrados:\n")
            print("ID             | Nome")
            for cliente in Clientes:
                    print(cliente['ID'] + " | " + cliente['Nome'])

        elif opcao == 3:
            clear_output()
            print("Descadastramento Iniciado\nDigite o ID do cliente a ser descadastrado:")
            id_remover = input("Digite o ID: ")

            cliente_encontrado = False
            nova_lista = []

            for cliente in Clientes:
                if cliente['ID'] == id_remover:
                    cliente_encontrado = True
                    continue
                nova_lista.append(cliente)

            if cliente_encontrado:
                with open(caminho_csv, mode='w', newline='', encoding='utf-8') as arquivo_csv:
                    campos = ['Data_Cadastro', 'ID', 'Nome', 'Pontos']
                    escritor = csv.DictWriter(arquivo_csv, fieldnames=campos, delimiter=';')
                    escritor.writeheader()
                    escritor.writerows(nova_lista)

                print("Cliente com ID", id_remover, "removido com sucesso.")
            else:
                print("Cliente com ID", id_remover, "não encontrado.")
        
        elif opcao == 0:
            clear_output()
            print("Saindo...")
            break
```

## Pedidos

```python
def Pedidos_():
    caminho_csv_pedidos = r"C:\Users\Desktop\Documents\Trabalho\Automação\Projeto\Auxiliar\Pedidos.csv"
    caminho_csv_cardapio = r"C:\Users\Desktop\Documents\Trabalho\Automação\Projeto\Auxiliar\Cardapio.csv"
    caminho_csv_clientes = r"C:\Users\Desktop\Documents\Trabalho\Automação\Projeto\Auxiliar\Clientes.csv"

    Cardapio = carregar_arquivo(caminho_csv_cardapio, ",")

    mapa_itens = {item['ID_Item']: item['Item'] for item in Cardapio}

    while True:
        Clientes = carregar_arquivo(caminho_csv_clientes, ";")
        Pedidos = carregar_arquivo(caminho_csv_pedidos, ";")

        print("Digite a opção desejada:"
            "\n1. Registrar um Pedido"
            "\n2. Ver Fila de Pedidos"
            "\n3. Fechar um Pedido"
            "\n0. Sair")
        
        try:
            opcao = int(input("Opção: "))
        except ValueError:
            clear_output()
            print("Digite um número válido.")
            input("Pressione Enter para continuar...")
            continue
        
        if (opcao == 1):
            itens = int(input("Quantos itens serão pedidos? "))
            cadastro = int(input("O Cliente possui cadastro?\n1. Sim\n2. Não\nOpção: "))

            codigo_pedido = datetime.now().strftime("%H%M%S")

            lista_nomes = []
            for i in range(itens):
                id_item = input("Qual o ID do item " + str(i+1) + "? ").strip()
                if id_item in mapa_itens:
                    lista_nomes.append(mapa_itens[id_item])
                else:
                    print("ID", id_item, "não encontrado no cardápio.")

            id_cliente = ""
            if cadastro == 1:
                id_cliente = input("Qual o ID do Cliente? ").strip()
                for cliente in Clientes:
                    if cliente['ID'] == id_cliente:
                        cliente['Pontos'] = int(cliente['Pontos']) + (itens * 10)
                        with open(caminho_csv_clientes, mode='w', newline='', encoding='utf-8') as arquivo_csv:
                            campos = ['Data_Cadastro', 'ID', 'Nome', 'Pontos']
                            escritor = csv.DictWriter(arquivo_csv, fieldnames=campos, delimiter=';')
                            escritor.writeheader()
                            escritor.writerows(Clientes)
                        break
                else:
                    print("Cliente não encontrado, pedido será registrado sem vínculo.")

            novo_pedido = {
                'Codigo': codigo_pedido,
                'Item': ",".join(lista_nomes),
                'Cliente': id_cliente
            }

            try:
                with open(caminho_csv_pedidos, mode='x', newline='', encoding='utf-8') as arquivo_csv:
                    campos = ['Codigo', 'Item', 'Cliente']
                    escritor = csv.DictWriter(arquivo_csv, fieldnames=campos, delimiter=';')
                    escritor.writeheader()
            except FileExistsError:
                pass

            with open(caminho_csv_pedidos, mode='a', newline='', encoding='utf-8') as arquivo_csv:
                campos = ['Codigo', 'Item', 'Cliente']
                escritor = csv.DictWriter(arquivo_csv, fieldnames=campos, delimiter=';')
                escritor.writerow(novo_pedido)
            clear_output()
            print("Pedido registrado com sucesso.")
            print("Código:", codigo_pedido)
            print("Itens:", ", ".join(lista_nomes))
            if id_cliente:
                print("Cliente:", id_cliente)
        
        elif (opcao == 2):
            clear_output()
            print("Pedidos:")
            print("Codigo | Pedido")
            for pedido in Pedidos:
                print(pedido['Codigo'] + " | " + pedido['Item'])
            print("\n")

        elif (opcao == 3):
            clear_output()
            print("Fechar Pedido\nDigite o código do pedido a ser fechado:")
            codigo_remover = input("Código: ").strip()

            pedido_encontrado = False
            nova_lista = []

            for pedido in Pedidos:
                if pedido['Codigo'] == codigo_remover:
                    pedido_encontrado = True
                    continue
                nova_lista.append(pedido)

            if pedido_encontrado:
                with open(caminho_csv_pedidos, mode='w', newline='', encoding='utf-8') as arquivo_csv:
                    campos = ['Codigo', 'Item', 'Cliente']
                    escritor = csv.DictWriter(arquivo_csv, fieldnames=campos, delimiter=';')
                    escritor.writeheader()
                    escritor.writerows(nova_lista)
                print("Pedido com código", codigo_remover, "fechado com sucesso.")
            else:
                print("Pedido com código", codigo_remover, "não encontrado.")
        
        elif opcao == 0:
            clear_output()
            print("Saindo...")
            break
```
